```python

import matplotlib.pyplot as plt
```

```python
#1
import numpy as np
N=100
ts=np.linspace(0,20,200)
ys=np.sin(ts)

def NUDFT (ts, ys, f):
    F=0j
    for k in range (len(ts)):
        F+= ys[k]*np.exp(-2*np.pi*1j*ts[k]*f)
    return F

NUDFT(ts,ys,0.1)


    

```

```python
#2
ts=np.linspace(0,20,200)
y2=np.sin(2*np.pi*np.sqrt(2)*ts)

NUDFT(ts,y2,0.1)
```

```python
dt=ts[1]-ts[0]
f=np.linspace(0,1/dt,100)
F=np.zeros(100)+0j

for k in range(len(f)):
   F[k]=NUDFT(ts,y2,f[k])

F
```

```python

plt.plot(f, abs(F), marker='o')  # marker='o' agrega puntos en cada valor
plt.title("Transformada de Fourier")  # Título del gráfico
plt.xlabel('X')  # Etiqueta del eje X
plt.ylabel('Y')  # Etiqueta del eje Y
plt.grid(True)  # Mostrar una cuadrícula
plt.show()  # Mostrar el gráfico
```

Podemos observar que se puede recuperar el valor de la frecuencia principal, es decir f=6.28 al añadir el ruido en la variable del tiempo. Al añadir ruido en los datos del tiempo se reduce el efecto alisasing, al añadir aleatoriedad a dichos puntos. Esto permite distinguir las frecuencias altas de las frecuencias bajas.  Al hacer esto es posible identificar frecuencias superiores a la frecuencia de Nyquist y es por esto que podemos ver más picos en nuestra transformada.
