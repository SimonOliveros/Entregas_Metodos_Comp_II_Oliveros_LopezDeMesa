```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

```python
N = 100
x= np.linspace(-1.5,1.5,N)
y = np.linspace(-1.5,1.5,N)

U = np.random.rand(N,N)
for i in range(N):
    for j in range(N):
        if np.abs(np.sqrt(x[i]**2+y[j]**2))>=1:
            theta=np.atan2(x[i],y[j])
            U[i,j]=np.sin(7*theta)
        

plt.imshow(U, extent=[-1.5, 1.5, -1.5, 1.5], origin='upper', norm='log')
plt.colorbar()
plt.show()
```

```python
N = 100
x= np.linspace(-1.5,1.5,N)
y = np.linspace(-1.5,1.5,N)

U = np.random.rand(N,N)
h=x[1]-x[0]
for i in range(N):
    for j in range(N):
        if np.abs(np.sqrt(x[i]**2+y[j]**2))>=1:
            theta=np.atan2(x[i],y[j])
            U[i,j]=np.sin(7*theta)
        if np.abs(np.sqrt(x[i]**2+y[j]**2))<=1:
            U[i,j]=(1/4)*(U[i,j+1]+U[i,j-1]+U[i-1,j]+U[1+i,j])-np.pi*(x[i]+y[j])*h**2

plt.imshow(U, extent=[-1.5, 1.5, -1.5, 1.5], origin='upper', norm='log')
plt.colorbar()
plt.show()
```
